# Supervised Learning (Regresi)

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
train_df = pd.read_csv("dataset/train.csv")
train_df
```

```python
test_df = pd.read_csv('dataset/test.csv')
test_df
```

```python
train_df.info()
```

```python
train_df.describe(include="all")
```

```python
missing_values = train_df.isna().sum()
missing_values[missing_values > 0]
```

```python
for feature in train_df.columns:
    plt.figure(figsize=(10,6))
    sns.boxplot(x=train_df[feature])
    plt.title(f'Boxplot of {feature}')
    plt.show()
```

```python
# Hapus Outlier
Q1 = train_df.quantile(0.25)
Q3 = train_df.quantile(0.75)
IQR = Q3 - Q1

# Filter baris yang tidak mengandung outlier
condition = ~((train_df < (Q1 - 1.5 * IQR)) | (train_df > (Q3 + 1.5 * IQR))).any(axis=1)
df = train_df.loc[condition, train_df.columns]
```

```python
for feature in df.columns:
    plt.figure(figsize=(10,6))
    sns.boxplot(x=df[feature])
    plt.title(f'Boxplot of {feature}')
    plt.show()
```

```python
# Mendapatkan kolom numerik saja
numeric_features = df.select_dtypes(include=[np.number]).columns
numeric_features
```

```python
# Normalisasi data numeric
scaler = StandardScaler()
df[numeric_features] = scaler.fit_transform(df[numeric_features])
```

```python
# Indentifikasi Baris duplikat
duplicates = df.duplicated()

print("Duplicated", df[duplicates])
```

```python
df.describe(include="all")
```

```python
# Menghitung jumlah variable
num_vars = df.shape[1]

# Menentukan jumlah baris untuk subplot
n_cols = 4
n_rows = -(-num_vars)

# Membuat subplot
fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, n_rows * 4))

# Flatten axes
axes = axes.flatten()

# Plot untuk setiap variable
for i, column in enumerate(df.drop(columns=["id"]).columns):
    df[column].hist(ax=axes[i], bins=20, edgecolor='black')
    axes[i].set_title(column)
    axes[i].set_xlabel('Value')
    axes[i].set_ylabel('Frequency')

# Menghapus subplot yang tidak terpakai (jika ada)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Sesuaikan layout
plt.tight_layout()
plt.show()
```
